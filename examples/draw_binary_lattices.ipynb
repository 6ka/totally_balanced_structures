## Totally balanced structure creation

First we create a dismantlable lattice. It is then binarized. We finally add new elements in order to be atomistic.

### dismantlable lattice

```python
# tbs imports
import tbs.dismantlable
from tbs.gamma_free import GammaFree
```

```python
NUMBER_LATTICE_ELEMENTS = 20
```

```python
lattice = tbs.dismantlable.random_dismantlable_lattice(range(NUMBER_LATTICE_ELEMENTS))
```

### binarization

```python
lattice.binarize({lattice.top, lattice.bottom})
```

### atomic lattice

```python
lattice.make_atomistic()
```

## Context Matrix

```python
context_matrix = GammaFree.from_lattice(lattice)
context_matrix.reorder_doubly_lexical()
```

### Cross table

```python
print(context_matrix)
```

### Clusters

```python
import tbs.gamma_free
print(tbs.gamma_free.to_string(context_matrix))
```

## cover graph
We will draw each element accordind to it's hierarchical depth.

We will draw it using matplotlib and each element will be represented by:
  - a triangle facing up if it's an element (sup irreducible)
  - a triangle facing down if it's an attribute (inf irreducible)
  - a losange if it is both an element and an attribute (doubly irreducible)
  - a circle otherwise


The edges will be plain and with the same color if it links two elements from the same hierarchy and dashed otehrwise.

```python
import matplotlib
from tbs.gamma_free import draw_to_pyplot
```

Draw a doubly lexically ordered context matrix

```python
matplotlib.pyplot.figure(figsize=(20, 10))

draw_to_pyplot(context_matrix, matplotlib.pyplot, matplotlib.cm.rainbow)
matplotlib.pyplot.show()
```

## Hierarchical ordering

```python
hierarchies = lattice.hierarchical_decomposition()
print(hierarchies)
```

hierarchical reordering according to a DLO order. This ensure that the order is compatible for each hierarchy

```python
new_order_lists = [[]]
for x in context_matrix.elements:  
    if hierarchies[x] < len(new_order_lists):
        new_order_lists[hierarchies[x]].append(x)
    else:
        new_order_lists.append([x])
        
new_order = [u for y in new_order_lists for u in y]
```

```python
print(new_order_lists)
```

```python
context_matrix.reorder_elements(new_order)
context_matrix.reorder_gamma_free_from_strongly_chordal_element_order()
```

```python
print(context_matrix.is_gamma_free())
print(context_matrix.is_doubly_lexically_ordered())
```

```python
matplotlib.pyplot.figure(figsize=(20, 10))

draw_to_pyplot(context_matrix, matplotlib.pyplot, matplotlib.cm.rainbow)
matplotlib.pyplot.show()
```
